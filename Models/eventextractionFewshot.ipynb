```python
import pandas as pd
from langchain_core.prompts.few_shot import FewShotPromptTemplate
from langchain_core.prompts.prompt import PromptTemplate
import ollama
from langchain.llms import Ollama
from langchain.chains import LLMChain
import json
import requests
from langchain.llms import BaseLLM
```

```python
ollama = Ollama(model="llama2fewshottest4", base_url="http://localhost:11434")
ollama("en log in ur acc, also nothing really changes in my opinion still causes some bugs and sudden exit in the app, other one is when u post a myday even your are done is sudenly it just go agin in the posting part like choosing the pic u want to post.")
```

```python
def predict_event(reviews):
    events = []
    for review in reviews:
        event = ollama(review)
        events.append(event)
    return events
```

```python
df = pd.read_csv('../Training Dataset/FewShotForEvents.csv')
dfReviews = df['content'].tolist()
dfReviews
```

```python
zeroShotEvents = predict_event(dfReviews)
```

```python
zeroShotEvents
```

```python
reviews_df = pd.DataFrame({'Review': dfReviews})

# Create DataFrame for events
events_df = pd.DataFrame({'Event': zeroShotEvents})

# Concatenate the DataFrames
fewshotEvent_df = pd.concat([reviews_df, events_df], axis=1)
```

```python
fewshotEvent_df.to_csv('../Generated Dataset/Keywords Events/LLAMA2/EventsGeneratedForKeywords.csv', index=False)
```

```python
df = pd.read_csv('../Generated Dataset/Keywords Events/LLAMA2/EventsGeneratedForKeywords.csv')
df.head(10)
```

```python

# Function to remove duplicates while preserving order
def remove_duplicates(events_text):
    # Split the events by line
    events = events_text.split('\n')
    seen = set()
    unique_events = []
    for event in events:
        if event not in seen:
            seen.add(event)
            unique_events.append(event)
    return '\n'.join(unique_events)

# Apply function to 'Event' column
df['Event'] = df['Event'].apply(remove_duplicates)
```

```python
df.to_csv('../Generated Dataset/Keywords Events/LLAMA2/EventsGeneratedForKeywords2.csv', index=False)

```
