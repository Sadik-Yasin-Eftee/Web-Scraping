```python
keyPhrase = ['after', 'as soon as', 'before', 'everytime', 'then', 'until', 'when', 'whenever', 'while']
```

```python
facebookReview = pd.read_csv('../App Reviews/Data/Facebook.csv')
facebookReview.head(10)
```

```python
instgramReview = pd.read_csv('../App Reviews/Data/Instagram.csv')
instgramReview.head(10)
```

```python
linkedinReview = pd.read_csv('../App Reviews/Data/LinkedIn.csv')
linkedinReview.head(10)
```

```python
tiktokReview = pd.read_csv('../App Reviews/Data/TikTok.csv')
tiktokReview.head(10)
```

```python
twitterReview = pd.read_csv('../App Reviews/Data/Twitter.csv')
twitterReview.head(10)
```

```python
whatsappReview = pd.read_csv('../App Reviews/Data/WhatsApp.csv')
whatsappReview.head(10)
```

```python
df = pd.concat([facebookReview, instgramReview, linkedinReview, tiktokReview, twitterReview, whatsappReview], ignore_index= True)
df.shape
```

```python
keywordsDataframe = df[df['content'].str.contains('|'.join(keyPhrase))]
keywordsDataframe.head(10)
keywordsDataframe.shape
```

```python
keywordsDataframe.drop(columns=['reviewId', 'score'], inplace=True)
keywordsDataframe.shape
keywordsDataframe.head(10)
```

```python
keywordsDataframe.reset_index(drop=True, inplace=True)
keywordsDataframe.head(10)
```

```python
keywordsDataframe['content'].loc[0]
```

```python
import re

def filter_single_sentence(review):
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', review)  
    return len(sentences) == 1  

single_sentence_reviews = keywordsDataframe[keywordsDataframe['content'].apply(lambda x: filter_single_sentence(x))]
single_sentence_reviews.head(10)
```

```python
single_sentence_reviews.reset_index(drop=True, inplace=True)
single_sentence_reviews.head(10)
```

```python
single_sentence_reviews['content'].loc[0]
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def extract_events(review_text):
    doc = nlp(review_text)
    events = []
    for sent in doc.sents:
        for token in sent:
            if token.dep_ in ['ROOT', 'advcl', 'conj'] and token.pos_ == 'VERB':
                event_phrase = token.subtree
                event_text = ' '.join([t.text.strip() for t in event_phrase if not t.is_punct])
                events.append(event_text)
    return events
```

```python
single_sentence_reviews['events'] = single_sentence_reviews['content'].apply(extract_events)
single_sentence_reviews.head(10)
```
